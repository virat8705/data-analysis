```python
import matplotlib.pyplot as plt
```

```python
x=[2,4,6,8,10]
y=[5,7,4,2,9]

plt.plot(x,y,color='r')

plt.fill_between(x=[6,8],y1=3,y2=6)

plt.title("python")
plt.xlabel("x-axis")
plt.ylabel('y-axis')
plt.show()
```

```python
import numpy as np
x=np.array([2,4,6,8,10])
y=np.array([5,7,4,2,9])

plt.plot(x,y,color='r')

plt.fill_between(x,y,color='c',where=(x>3) & (x<7))

plt.title("python")
plt.xlabel("x-axis")
plt.ylabel('y-axis')
plt.show()
```

```python
import numpy as np
x=np.array([2,4,6,8,10])
y=np.array([5,7,4,2,9])

plt.plot(x,y,color='r')

plt.fill_between(x,y,color='c',where=(x>3) & (x<7),alpha=0.5)

plt.title("python")
plt.xlabel("x-axis")
plt.ylabel('y-axis')
plt.show()
```
