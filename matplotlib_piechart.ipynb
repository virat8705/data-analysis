```python
import matplotlib.pyplot as plt
```

```python
x= [33,30,10,37]
y= ['c','c++','Java','python']
ex = [0.0,0.0,0.2,0.0]
c = ['r','y','g','m']
plt.pie(x,labels = y,explode = ex,colors = c,autopct = '%0.3f%%')


plt.show()
```

```python
x= [33,30,10,37]
y= ['c','c++','Java','python']
ex = [0.0,0.0,0.2,0.0]
c = ['r','y','g','m']
plt.pie(x,labels = y,explode = ex,colors = c,autopct = '%0.2f%%',shadow=True,radius = 1.3,
       textprops = {'fontsize': 15},counterclock=False)


plt.show()
```

```python
x= [33,30,10,37]
y= ['c','c++','Java','python']
ex = [0.0,0.0,0.2,0.0]
c = ['r','y','g','m']
plt.pie(x,labels = y,explode = ex,colors = c,autopct = '%0.2f%%',shadow=True,radius = 1,
       textprops = {'fontsize': 10},counterclock=False,rotatelabels=False)

plt.title('Piechart')
plt.legend(loc = 3)
plt.show()
```
