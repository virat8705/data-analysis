```python
import pandas as pd
```

```python
import statsmodels.api as sm
```

```python
df = pd.read_csv('Advertising.csv',index_col=0)
```

```python
df
```

```python
x = df[['TV','radio','newspaper']]
```

```python
y=df['sales']
```

```python
x = sm.add_constant(x)
```

```python
x
```

```python
model = sm.OLS(y,x).fit()
```

```python
model.summary()
```
