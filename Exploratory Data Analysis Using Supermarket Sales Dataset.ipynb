#### Display Daily Sales by Day of the Week

```python
data.columns
```

```python
dw_mpping={
    0:'Mon',
    1:'Tue',
    2:'Wed',
    3:'Thur',
    4:'Fri',
    5:'Sat',
    6:'Sun'
}
```

```python
data['day_of_week']=data['Date'].dt.dayofweek.map(dw_mpping)
```

```python
data['day_of_week'].value_counts().plot(kind='bar')
```

#### What Will Be the Highest Months for Sales

```python
 month_map={
     1:'Jan',
     2:'Feb',
     3:'Mar',
     4:'Apr',
     5:'May',
     6:'Jun',
     7:'Jul',
     8:'Aug',
     9:'Sep',
     10:'Oct',
     11:'Nov',
     12:'Dec'
 }
```

```python
data['month_name']=data['Date'].dt.month.map(month_map)
```
