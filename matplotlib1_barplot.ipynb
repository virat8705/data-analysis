```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=['Python','C++','C','Java']
y=[85,60,72,82]
plt.xlabel("language")
plt.ylabel("Numbers")
plt.title("Bar Graph")
c=['y','g','b','m']
plt.bar(x,y,width=0.8,color=c)
plt.show()
```

```python
x=['Python','C++','C','Java']
y=[85,60,72,82]
plt.xlabel("language")
plt.ylabel("Numbers")
plt.title("Bar Graph")
c=['y','g','b','m']
plt.bar(x,y,width=0.4,color=c,edgecolor='r',linewidth=3,linestyle=':',alpha=0.5)
plt.show()
```

```python
x=['Python','C++','C','Java']
y=[85,60,72,82]
plt.xlabel("language")
plt.ylabel("Numbers")
plt.title("Bar Graph")
c=['y','g','b','m']
plt.bar(x,y,width=0.4,color=c,edgecolor='r',label='popularity')
plt.legend()
plt.show()
```

```python
x=['Python','C++','C','Java']
y=[85,60,72,82]
z=[20,30,40,50]
p=[0,1,2,3]
plt.xlabel("language",fontsize=15)
plt.ylabel("Numbers",fontsize=15)
plt.title("Bar Graph",fontsize=15)
c=['y','g','b','m']
width=0.3
plt.bar(x,y,width,color='r',edgecolor='r',label='popularity')
plt.bar(x,z,width,color='y',edgecolor='r',label='popularity1')
plt.legend()
plt.show()
```

```python
x=['Python','C++','C','Java']
y=[85,60,72,82]
z=[20,30,40,50]

width=0.3
p=np.arange(len(x))
p1=[j+width for j in p]

plt.xlabel("language",fontsize=15)
plt.ylabel("Numbers",fontsize=15)
plt.title("Bar Graph",fontsize=15)


plt.bar(p,y,width,color='r',edgecolor='r',label='popularity')
plt.bar(p1,z,width,color='y',edgecolor='r',label='popularity1')

plt.xticks(p+width/2,x,rotation=75)
plt.legend()
plt.show()
```

```python
x=['Python','C++','C','Java']
y=[85,60,72,82]
z=[20,30,40,50]

width=5.3
p=np.arange(len(y))
p1=[j+width for j in p]

plt.xlabel("language",fontsize=15)
plt.ylabel("Numbers",fontsize=15)
plt.title("Bar Graph",fontsize=15)


plt.barh(y,p,width,color='r',edgecolor='r',label='popularity')
plt.barh(z,p1,width,color='y',edgecolor='r',label='popularity1')

plt.xticks(p+width/2,x,rotation=75)
plt.legend()
plt.show()
```
