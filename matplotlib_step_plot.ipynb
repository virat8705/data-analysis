```python
import matplotlib.pyplot as plt
```

```python
x=[3,5,2,8,4]
y=[6,2,7,3,8]

plt.step(x,y)


plt.title('python')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.grid()
plt.show()
```

```python
x=[3,5,2,8,4]
y=[6,2,7,3,8]

plt.step(x,y,color='r',marker='o')


plt.title('python')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.grid()
plt.show()
```

```python
x=[3,5,2,8,4]
y=[6,2,7,3,8]

plt.step(x,y,color='r',marker='o',ms=10,mfc='g',label='python')

plt.legend(loc=2)
plt.title('python')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.grid()
plt.show()
```
