```python
import matplotlib.pyplot as plt
```

```python
day=[1,2,3,4,5,6,7]
no=[2,1,5,7,8,9,3]
plt.scatter(day,no,color='r')
plt.title("Scatter plot",fontsize=15)
plt.xlabel("Day",fontsize=15)
plt.ylabel("No.",fontsize=15)
plt.show()
```

```python
day=[1,2,3,4,5,6,7]
no=[2,1,5,7,8,9,3]
color = ['r','y','g','b','c','m','r']
size = [130,135,150,170,130,140,125]
plt.scatter(day,no,c = color,s=size,alpha=0.5)
plt.title("Scatter plot",fontsize=15)
plt.xlabel("Day",fontsize=15)
plt.ylabel("No.",fontsize=15)
plt.show()
```

```python
day=[1,2,3,4,5,6,7]
no=[2,1,5,7,8,9,3]
color = ['r','y','g','b','c','m','r']
size = [230,325,520,270,230,240,225]
plt.scatter(day,no,c = color,s=size,alpha=0.9,marker='X',edgecolor="k",linewidth = 2)
plt.title("Scatter plot",fontsize=15)
plt.xlabel("Day",fontsize=15)
plt.ylabel("No.",fontsize=15)
plt.show()
```

```python
day=[1,2,3,4,5,6,7]
no=[2,1,5,7,8,9,3]
color = [50,38,14,26,44,56,66]
size = [130,135,150,170,130,140,125]
plt.scatter(day,no,c = color,s=size,cmap = 'viridis',alpha = 0.8)
plt.colorbar().set_label("ColorBar",fontsize = 15)
plt.title("Scatter plot",fontsize=15)
plt.xlabel("Day",fontsize=15)
plt.ylabel("No.",fontsize=15)
plt.show()
```

```python
day=[1,2,3,4,5,6,7]
no=[2,1,5,7,8,9,3]
no2 = [5,3,8,6,2,5,9]
color = [50,38,14,26,44,56,66]
size = [130,135,150,170,130,140,125]
plt.scatter(day,no,c = color,s=size,cmap = 'viridis',alpha = 0.8)
plt.scatter(day,no2,c = color,s=size,cmap = 'Oranges',alpha = 0.8)
plt.colorbar().set_label("ColorBar",fontsize = 15)
plt.title("Scatter plot",fontsize=15)
plt.xlabel("Day",fontsize=15)
plt.ylabel("No.",fontsize=15)
plt.show()
```
