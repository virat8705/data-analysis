```python
import matplotlib.pyplot as plt
```

```python
x=[10,20,30,40,50,60,70]
plt.boxplot(x)
plt.show()
```

```python
x=[10,20,30,40,50,60,70]
plt.boxplot(x,notch=True)
plt.show()
```

```python
x=[10,20,30,40,50,60,70]
plt.boxplot(x,vert = False)
plt.show()
```

```python
x=[10,20,30,40,50,60,70]
plt.boxplot(x,widths = 0.5)
plt.show()
```

```python
x=[10,20,30,40,50,60,70]
plt.boxplot(x,labels=['python'])
plt.show()
```

```python
x=[10,20,30,40,50,60,70,120,200]
plt.boxplot(x,labels=['python'],patch_artist=True,
            showmeans=True)
plt.show()
```

```python
x=[10,20,30,40,50,60,70,120,200]
plt.boxplot(x,labels=['python'],patch_artist=True,
            showmeans=True,whis = 4)
plt.show()
```

```python
x=[10,20,30,40,50,60,70,120,200]
plt.boxplot(x,labels=['python'],patch_artist=True,
            showmeans=True,sym='g+')
plt.show()
```

```python
x=[10,20,30,40,50,60,70,120,200]
plt.boxplot(x,labels=['python'],
            showmeans=True,sym='g+',boxprops=dict(color='r'),capprops=dict(color='r'),
           whiskerprops=dict(color='c'))
plt.show()
```

```python
x=[10,20,30,40,50,60,70,120]
plt.boxplot(x,labels=['python'],
            showmeans=True,sym='g+',boxprops=dict(color='r'),capprops=dict(color='r'),
           whiskerprops=dict(color='c'),flierprops=dict(markeredgecolor='y'))
plt.show()
```

```python
x=[10,20,30,40,50,60,70]
x1=[10,20,40,50,30,60,90]
y = [x,x1]

plt.boxplot(y,labels=['python','c++'],showmeans=True,sym='g+',boxprops=dict(color='r'),capprops=dict(color='r'),
           whiskerprops=dict(color='c'))
plt.show()
```
