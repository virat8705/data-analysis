```python
data['Size'] = data['Size'].str.replace('Varies with device', '0')
```

```python
data.head()
```

```python
def convert_to_mb(size_str):
    size_str = size_str.strip()
    if size_str.endswith('M'):
        return float(size_str.rstrip('M'))
    elif size_str.endswith('K'):
        return float(size_str.rstrip('K')) / 1024
    else:
        return 0
```

```python
data['Size'] = data['Size'].apply(convert_to_mb)
```

```python
data['Size'].unique()
```

```python
data['Installs'] = data['Installs'].str.replace('[+,]', '', regex=True).astype(int)
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.impute import SimpleImputer
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
```

```python
features = data[['Reviews', 'Size', 'Installs']]
target = data['Rating']
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)

missing_target_rows = np.isnan(y_train)
X_train = X_train[~missing_target_rows]
y_train = y_train[~missing_target_rows]

imputer = SimpleImputer(strategy='mean')
y_train = imputer.fit_transform(y_train.values.reshape(-1, 1)).flatten()
y_test = imputer.fit_transform(y_test.values.reshape(-1, 1)).flatten()

model = LinearRegression()
model.fit(X_train, y_train)

predictions = model.predict(X_test)

mse = mean_squared_error(y_test, predictions)
rmse = np.sqrt(mse)
mae = mean_absolute_error(y_test, predictions)
r2 = r2_score(y_test, predictions)
print(f"Mean Squared Error: {mse}")
print(f"Root Mean Squared Error (RMSE): {rmse:.2f}")
print(f"Mean Absolute Error (MAE): {mae:.2f}")
print(f"R-squared (R^2): {r2:.2f}")

```

```python
from sklearn.ensemble import RandomForestRegressor

features = data[['Reviews', 'Size', 'Installs']]
target = data['Rating']

X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)

missing_target_rows = np.isnan(y_train)
X_train = X_train[~missing_target_rows]
y_train = y_train[~missing_target_rows]

imputer = SimpleImputer(strategy='mean')
y_train = imputer.fit_transform(y_train.values.reshape(-1, 1)).flatten()
y_test = imputer.fit_transform(y_test.values.reshape(-1, 1)).flatten()

model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

predictions = model.predict(X_test)

mse = mean_squared_error(y_test, predictions)
rmse = np.sqrt(mse)
mae = mean_absolute_error(y_test, predictions)
r2 = r2_score(y_test, predictions)
print(f"Mean Squared Error: {mse}")
print(f"Root Mean Squared Error (RMSE): {rmse:.2f}")
print(f"Mean Absolute Error (MAE): {mae:.2f}")
print(f"R-squared (R^2): {r2:.2f}")
```
