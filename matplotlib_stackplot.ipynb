```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5]
area = [4,2,6,3,8]
        

plt.stackplot(x,area)

plt.show()
```

```python
x = [1,2,3,4,5]
area = [4,2,6,3,8]
area1 = [5,7,6,4,9]
area2 = [6,3,4,7,6]        

l=['area1','area2','area3']
plt.stackplot(x,area,area1,area2,labels=l)

plt.legend(loc = 2)
plt.show()
```

```python
x = [1,2,3,4,5]
area = [4,2,6,3,8]
area1 = [5,7,6,4,9]
area2 = [6,3,4,7,6]        

l=['area1','area2','area3']
plt.stackplot(x,area,area1,area2,labels=l,colors=['y','c','m'])

plt.legend(loc = 2)
plt.show()
```

```python
x = [1,2,3,4,5]
area = [4,2,6,3,8]
area1 = [5,7,6,4,9]
area2 = [6,3,4,7,6]        

l=['area1','area2','area3']
plt.stackplot(x,area,area1,area2,labels=l,colors=['y','c','m'])

plt.legend(loc = 2)
plt.savefig('line1',dpi=2000,facecolor='c')
plt.show()
```

```python
x = [1,2,3,4,5]
area = [4,2,6,3,8]
area1 = [5,7,6,4,9]
area2 = [6,3,4,7,6]        

l=['area1','area2','area3']
plt.stackplot(x,area,area1,area2,labels=l,colors=['y','c','m'])

plt.title("Python")
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend(loc = 2)
plt.show()
```
