```python
import matplotlib.pyplot as plt
```

```python
x=[1,5,4,3,7]
y=[5,4,2,7,6]

plt.stem(x,y)

plt.show()
```

```python
x=[1,5,4,3,7]
y=[5,4,2,7,6]

plt.stem(x,y,linefmt=":")

plt.show()
```

```python
x=[1,5,4,3,7]
y=[5,4,2,7,6]

plt.stem(x,y,linefmt=":",markerfmt='r+')

plt.show() 
```

```python
x=[1,5,4,3,7]
y=[5,4,2,7,6]

plt.stem(x,y,linefmt=":",markerfmt='r+',bottom = 2)

plt.show() 
```

```python
x=[1,5,4,3,7]
y=[5,4,2,7,6]

plt.stem(x,y,linefmt=":",markerfmt='r+',basefmt = 'g')

plt.show() 
```

```python
x=[1,5,4,3,7]
y=[5,4,2,7,6]

plt.stem(x,y,linefmt=":",markerfmt='r+',basefmt = 'g',label='python',use_line_collection = False)
plt.legend()

plt.show() 
```

```python
x=[1,5,4,3,7]
y=[5,4,2,7,6]

plt.stem(x,y,linefmt=":",markerfmt='r+',basefmt = 'g',label='python',use_line_collection = False,
        orientation = 'horizontal')
plt.legend()

plt.savefig('line3',dpi=2000,facecolor='c')
plt.show() 
```
